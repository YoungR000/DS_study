# load ex1 using tf.keras


```python
# adaline using keras
from tensorflow.keras import models, layers, optimizers, losses
import numpy as np

X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

T = np.array([0,1,1,0])

model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='sigmoid',input_shape=(2,))) 
# unit : layer에 있는 node의 개수, 

model.compile(optimizer='adam',
             loss="binary_crossentropy",
             metrics = ['acc'])

model.fit(X,T,epochs = 100000, batch_size = 1, verbose = 0)
```

```python
# test
Xtest_loss,Xtest_acc = model.evaluate(X,T)
print(Xtest_acc)
Ttest = model.predict(X)
print(Ttest)
```

#### 1) ISIR 데이터셋 들고 오시고
#### 2) train/test 나누시고
#### 3) 입력 4 -> 출력 1개인 Perceptron을 만드시고
#### 4) sgd / MSE를 써서 학습하고
#### 5) test 로 검증

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers
import numpy as np

#1) ISIR 데이터셋
X,y = load_iris().data, load_iris().target

#2) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size = 1/3)

#3) Perceptron model 을 만들겠습니다
model = models.Sequential()
model.add(layers.Dense(units = 1, activation = 'sigmoid', input_shape = X_train[1].shape))

#4) 학습 파라미터 설정
model.compile(optimizer = 'sgd',
             loss = 'mse',
             metrics = 'acc')

#5) 학습 시작
model.fit(X_train, y_train, epochs = 1000, batch_size=1)

Xtest_loss,Xtest_acc = model.evaluate(X_test,y_test)
print(Xtest_acc)
```

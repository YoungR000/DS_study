# 1. 책
1. 파이썬 머신러닝 완벽 가이드(권철민)
  * 머신러닝 알고리즘에 대한 전반적인 내용
  * 간단한 원리 설명 + 코드 구현 수록
  * 목차: 파이썬기초(넘파이/판다스), 사이킷런 기초, 평가(Metric), 분류, 회귀, 차원축소, 군집화, 텍스트분석, 추천시스템
  * 수록 코드 깃허브: https://github.com/wikibook/pymldg-rev
  * 인프런에 저자 강의도 있음(https://www.inflearn.com/course/파이썬-머신러닝-완벽가이드)
  
  
2. 밑바닥부터 시작하는 딥러닝(사이토 고키)
  * 총 3권으로 구성
    * 1권: 신경망 기초(퍼셉트론, 활성화함수, optimizer, 오차역전파 등), CNN 일부
    * 2권: RNN, 자연어처리 위주
    * 3권: 신경망 고도화
  * 딥러닝 활용에 필요한 기초지식들


3. 데이터 과학자와 데이터 엔지니어를 위한 인터뷰 문답집(Hulu데이터과학팀)
  * 머신러닝, 딥러닝, 강화학습 등의 알고리즘에 대한 이론적 지식을 문답식으로 기술
  * 애매했던 개념들을 정리하는 데 도움

# 6. 자격증
1. ADsP(데이터분석 준전문가)
  * 한국데이터산업진흥원 주관
  * https://www.dataq.or.kr/www/sub/a_06.do
  * 데이터분야 자격증 중에서 제일 쉬움
  * 총 3과목 : 1,2과목은 자격증을 위한 그냥 암기부분. 3과목이 메인이며 통계분석 / 정형데이터마이닝 위주
  * 데이터관련 지식을 체계적으로 정리하는 용도로 따볼만 했음(과거형)
  * 빅데이터 분석기사가 생기고 나서는 ADP자격증 응시를 위한 응시자격용
  * 필기시험 only
  
  
2. 빅데이터분석기사
  * 데이터분야 유일한 기사 자격증
  * 한국데이터산업진흥원에서 시행
  * https://www.dataq.or.kr/www/sub/a_07.do
  * ADsP와 시행처가 같아서 겹치는 내용이 많음
  * 데이터관련 지식을 체계적으로 정리하는 용도로 ADsP 대신 따볼만 함
  * 필기시험 : ADsP내용 + 딥러닝, 비정형데이터 기초 정도
  * 실기시험 : 간단한 개념문제 + 간단한 작업형 3문제 + 실제데이터예측 1문제 (난이도 매우 쉬움)
  
  
3. DATA
  * YBM IT 주관
  * https://www.ybmit.com/data/data_info.jsp
  * 과목은 Python, Excel 두 과목, 급수 각각 1,2급
  * Python 1급 기준, 객관식 필기문제 15개, 작업형 실기문제 10개
  * 필기 출제내용은 ADP 내용과 비슷한 것 같음
  * 실기는 통계 문제가 대부분이며 머신러닝은 군집화나 PCA정도가 나옴
  * 생긴지 얼마 안된 자격증이라 존재여부 자체를 모르는 사람들이 많음(실효성 거의 없음)
 
 
4. ADP(데이터분석 전문가)
  * 한국데이터산업진흥원 주관
  * https://www.dataq.or.kr/www/sub/a_05.do
  * 아직 국내에 취득한 사람이 수백 명 내외
  * 필기시험 : 빅데이터분석기사 필기내용 + 데이터병렬처리 + 데이터시각화
  * 실기시험 : 합격률 2%대의 극악의 난이도. 합격 점수컷도 높고(75/100), 필기내용에서 커버하지 않은 문제도 막 튀어나옴
  * 취득 난이도에 비해 실효성 거의 없음
